```python
cars.columns = ['#','marca_carro','modelo_carro','ano_fabricacao','modelos_vendidos_em_milhares',
                'velocidade_maxima_km','percentual_vendas_por_marca','vendas_milhares_2018','preco$']
```

<h3>Percentual de vendas de acordo com a marca do veiculo</h3>

```python
cars['marca_carro'].value_counts('percentual_vendas_por_marca')
```

<h3>Exibindo grafico</h3>

```python
cars['marca_carro'].value_counts('percentual_vendas_por_marca').plot()
```

<h3>Grafico de barras sobre percentual de vendas</h3>

```python
cars['marca_carro'].value_counts('percentual_vendas_por_marca').plot(kind="bar",figsize=(18,5),grid=False,rot=0,color="green")
```

<h3>Carros vendidos em 2018 de acordo com a marca</h3>

```python
mais_vendidos_2018 = cars[['marca_carro','vendas_milhares_2018']].head(10).set_index('marca_carro').sort_values('vendas_milhares_2018', ascending = True)
mais_vendidos_2018.plot(kind="barh",figsize=(15,7),grid=False,color="darkred",legend=False)

plt.title('Carros mais vendidos em 2018')
plt.xlabel('Quantidade de vendas ("Em milhares")')
plt.show()
```

<h3>Relação Marca x Modelo</h3>

```python
crosstab_vg = pd.crosstab(cars['marca_carro'], cars['modelo_carro'])
crosstab_vg.head(10)
```

<h3>Percentual de vendas de acorodo marca e modelo<h/h3>

```python
crosstab_vg['percentual_vendas_por_marca'] = crosstab_vg.sum(axis=1)
crosstab_vg.head(10)
```
